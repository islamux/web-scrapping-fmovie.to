## Initial Crawling

```python
def web_crawler(url):
    """
        Hi, I am web_crawler func. I take url as argument.
        For this project, I will take most-watched page's url.
        I will go through 50 pages of most-watched movies/tvs whatever there is .
        
        I will parse them using beautifulsoup.
        And I will return a list of html divs that have all the movies/tvs/others.
        
        I have try and except implemented whilst paging, so if the requests throws error,
        I will skip that page and go to next one. Hence, final product may not have all the items.
    """
    print(web_crawler.__doc__)
    

    # We'll fetch 50 pages worth data
    counter=0
    pages = [i for i in range(1,51)]
    # Following list will have lists ofcontent of body tag fetched from data.
    movie_list = []
    
    # Go thourhg 50 pages to have abundance of data.
    for page in pages: 
#         Create payload for params that has pages 
        payload={'page':str(page)}
        try:
            res = req.get(url, params=payload)
        except:
#             skip the page that gives error but inform the user.
            print("The page {} raised error, so i've moved on to the page {}".format(page,page+1))
            continue
        else:
            counter +=1
            print("Successfull Crawling on page {}".format(page))
            pars = bs(res.content)
            body = pars.body            
            head = pars.head      
            
#         find only the divs with class movie-list
            x = body.find_all("div", attrs={'class':'movie-list'})
#     Find the internal list with movies only with item classes and append to movie_list
            movie_list.append(x[0].find_all('div', attrs ={'class':'item'}))
#     print the status
            print("Successfull addition of movie list from page {}".format(page))

        

#After gaining all 50 pages movies lets flatten the nested movie lists into one list
    flat_movie_list = [item for sublist in movie_list for item in sublist]
    #Lets print the total pages surfed
    print("After avoiding exceptions. The toal pages of successfull crawling are {}".format(counter))
    
    
    return flat_movie_list
```

```python
movie_list = web_crawler(most_watched_url)
```

# Secondary Crawling

Site provides extra info like imdb, site_ratings and so on in the movie watch page, so lets make a function to loop over each url in the db, then make request from each url and then extract info


## Nan Checker for Str


But first lets make nan checker function. Math.nan only checks nan for numeric types not string, so we have to workaround that logic

```python
'''
    I am "nan_checker_for_strings" func. I am a customize nan checker,
    to over come flaw of math.nan that only checks real number not string.
    
    if we convert a srting into float it will raise error
    but for nan values it won't raise error.
    
    Which means if error is raised the value is string and not a null, 
    Since we want to find if the value is null or not.
    
    so we can return False for the string as it is not null and rasises error
    while return true for nan
'''

def nan_checker_for_strings(val):
    
    print(nan_checker_for_strings.__doc__)
    #let use try except:
    try:
        # it wont raise error for null so its
        math.isnan(float(val))
    except:
        # if error, means its string so not a null hence return false
        return False
    
    else:
        return True
```

## Extract Info 

```python
def extract_info(url):
    res1 = req.get(url)
    pars =bs(res1.content)
    body =pars.body
    
# When applyng the functions with df
# It was found that requests was not able to provide the desired content 
# Hence try and except is implemented to request and parse the content in case of errors
#for loop is executed cause bothe queries can give error.
#                     body.find_all('div', attrs={'class':'info'})[0].find_all('dd'),body.find_all('dt')]
    inf = ""
    try:
        body.find_all('div', attrs={'class':'info'})
    except:
        inf=np.nan
    else:
        try:
            body.find_all('div', attrs={'class':'info'})[0].find_all('dl')
        except:
            inf=np.nan
        else:
            inf = body.find_all('div', attrs={'class':'info'})
            dls = inf[0].find_all('dl')
            #FInd children of dls     
            dds_0 = dls[0].find_all('dd')
            dds_1= dls[1].find_all('dd')
            dts_0 = dls[0].find_all('dt')
            dts_1= dls[1].find_all('dt')
    ##--------------------------------------------------------------------------------------------------------

#     dict to hold values
    info={}
    ##--------------------------------------------------------------------------------------------------------
    if nan_checker_for_strings(inf):
        #return nan error later we can deal with them individually.
        return np.nan
        

    else:
       #Find INfo from first dl
        for dt,dd in zip(dts_0,dds_0):
            key = dt.string.strip(":")
            val_list = []
            try:
                dd.children.string

            except:
                for child in dd.children:
                    if len(child.string.strip())<=1:
                        pass
                    else:
                        val_list.append(child.string.strip())
                info[key] = val_list
            else:
                info[key] = val_list

        ##--------------------------------------------------------------------------------------------------------

        #Find INfo from second dl

        for dt,dd in zip(dts_1,dds_1):
            key = dt.string.strip(":")
            val_list = []
            try:
                # The director info has only one child but others have more than one.So
                dd.children.string

            except:
                # IF more than one child, loop through children and append string to list
                for child in dd.children:
                    # Check if the children are white comma
                    if len(child.string.strip())<=1:
                        # If so pass
                        pass
                    else:
                        # ELse add the string to value_list
                        val_list.append(child.string.strip())
                info[key] = val_list
                
            # if no error
            else:
    #             print(dd.children.string)
                info[key] = val_list
        ##--------------------------------------------------------------------------------------------------------
        # Imdb Is in separate section so extract it
        imdb =inf[0].find_all('span',attrs={'class':'imdb'})[0]
        for lst in imdb.next_siblings:
            #Siblings are some white spaces as well so add checker.
            if str.isspace(lst.string):
                pass
            else:
                info[imdb.string] = lst.string

        ##--------------------------------------------------------------------------------------------------------

        # some values are in list of strings
        # Extract them as one string using join
        info['Rating'] = "/".join(map(str,info['Rating'][:2]))+" times"
        info['Release'] = "".join(map(str,info['Release']))
        info['Genre'] = ",".join(map(str,info['Genre']))
        info['Stars'] = ",".join(map(str,info['Stars']))
        info['Director'] = "".join(map(str,info['Director']))
        info['Country'] = ",".join(map(str,info['Country']))
    return info
```

Extract information on movies and tvs using extract_info func

```python
movie_information = movie_df.loc[:,"watch_link"].apply(extract_info)
tv_information = tv_df.loc[:,"watch_link"].apply(extract_info)
```

## Unpack  with  Function

The info received above are in dictionary form, json_normalize crashed the kernel so lets write function for unpacking and return a new df that has info

```python
def list_unpacker(lst):

    # First take the df and then genereate list of dictonarues
    informations = lst.copy().tolist()
    new_df = pd.DataFrame( columns=['Genre','Stars','IMDb','Director','Release','Country'])
# Create a separate list of values which will be added as individual series later on
    genre =[]
    stars =[]
    rating =[]
    imdb=[]
    director=[]
    release_date= []
    country=[]
     # Using pd.json_normalize kept on crashing kernel, so lets use loop and append method.
    
    for info in informations:
        if nan_checker_for_strings(info):
            # add nans to the rows of the columns in df
            genre.append(np.nan)
            stars.append(np.nan)
            rating.append(np.nan)
            imdb.append(np.nan)
            director.append(np.nan)
            release_date.append(np.nan)
            country.append(np.nan)
        else:
            # if the reutrn is not nan
            genre.append(info['Genre'])
            stars.append(info['Stars'])
            rating.append(info['Rating'])
            imdb.append(info['IMDb'])
            director.append(info['Director'])
            release_date.append(info['Release'])
            country.append(info['Country'])
           
    # add the above lists as series in new_df
    new_df['Genre'] = genre
    new_df['Stars'] = stars
    new_df['Rating'] = rating
    new_df['IMDb'] = imdb
    new_df['Director'] = director
    new_df['Release'] = release_date
    new_df['Country'] = country


    return new_df
```

```python
movie_info_df = list_unpacker(movie_information)
```

```python
tv_info_df = list_unpacker(tv_information)
```
